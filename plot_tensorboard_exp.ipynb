```python
import os
import pandas as pd
import matplotlib.pyplot as plt

csv_files= os.listdir("tensorboard_plots")
plt.figure(figsize=(7, 5))

for file in csv_files:
    if 'ppo' in file:
        model = "PPO"
    elif 'sac' in file:
        model = 'SAC'

    data = pd.read_csv(f"tensorboard_plots/{file}")

    x = data['episode_number']
    y = data['episode_reward']

    plt.plot(x[:20000], y[:20000], label=model)

plt.xlabel("Experiment Number")
plt.ylabel("Experiment Reward")
plt.grid(True)
plt.legend()
    
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Sample data: substrate-to-cell ratios
substrate_to_cell_ratios = np.array([3900, 4100, 4000, 3950, 4200, 3980, 4050, 3800])
optimum_ratio = 4000  # Optimum substrate-to-cell ratio

# Calculate the RMSE
rmse = np.sqrt(np.mean((substrate_to_cell_ratios - optimum_ratio) ** 2))

# Plot the substrate-to-cell ratios
plt.figure(figsize=(8, 6))

# Scatter plot for observed ratios
plt.scatter(range(len(substrate_to_cell_ratios)), substrate_to_cell_ratios, color='blue', label='Substrate-to-Cell Ratio')

# Plot the optimum ratio (horizontal line)
plt.axhline(optimum_ratio, color='red', linestyle='--', label=f'Optimum Ratio: {optimum_ratio}')

# Show deviations from optimum as vertical lines
for i, ratio in enumerate(substrate_to_cell_ratios):
    plt.plot([i, i], [ratio, optimum_ratio], color='gray', linestyle='-', linewidth=1)  # Deviation lines

# Add RMSE as text on the plot
plt.text(0, max(substrate_to_cell_ratios) + 50, f'RMSE = {rmse:.3f}', fontsize=12, color='black')

# Labels and title
plt.xlabel("Experiment Number")
plt.ylabel("Substrate-to-Cell Ratio")
plt.title("Substrate-to-Cell Ratio and RMSE Visualization")
plt.legend()
#plt.grid(True)

# Show plot
plt.show()
```
