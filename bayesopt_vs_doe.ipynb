```python
import matplotlib.pyplot as plt
import numpy as np

# Data
methods = ["BBD",'CCD', 'Full Factorial', 'BayesOpt']
runs = [15,22, 125, 1000]
enzyme_activity_doe = [2.7, 2.9, 3.1,0]  # Max activity from DOE methods
enzyme_activity_bayesopt = [2.7, 2.8, 3.16, 3.099]  # Max activity from BayesOpt

# Create bar positions
x = np.arange(len(methods))  # the label locations
width = 0.35  # the width of the bars

# Plotting
fig, ax = plt.subplots()
bars1 = ax.bar(x - width/2, enzyme_activity_doe, width, label='DOE')
bars2 = ax.bar(x + width/2, enzyme_activity_bayesopt, width, label='BayesOpt')

# Labels and Titles
ax.set_xlabel('Methods')
ax.set_ylabel('Max Enzyme Activity (U/L)')
ax.set_xticks(x)
ax.set_xticklabels([f'{method} ({run} runs)' for method, run in zip(methods, runs)])
ax.legend()

# Display values on bars
for bar in bars1 + bars2:
    height = bar.get_height()
    ax.annotate(f'{height:.1f}',
                xy=(bar.get_x() + bar.get_width() / 2, height),
                xytext=(0, 3),  # 3 points vertical offset
                textcoords="offset points",
                ha='center', va='bottom')

plt.tight_layout()

plt.show()
```
