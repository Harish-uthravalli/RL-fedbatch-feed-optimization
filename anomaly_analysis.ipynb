```python
import pandas as pd

df1 = pd.read_csv("rl_data.csv")
df2 = pd.read_csv("rl2_data.csv")
```

```python
import matplotlib.pyplot as plt
x = df1['timestep']/100
y = df1['delE']

x2 = df2['timestep']/100
y2 = df2['delE']

plt.plot(x, y, label="MuX=0.1")
plt.plot(x2, y2,label="MuX=0.2")
plt.xlabel("Time")
plt.ylabel("Enzymes Produced")
plt.legend()
```

```python
import matplotlib.pyplot as plt
x = df1['timestep']/100
y = df1['dxdt']

x2 = df2['timestep']/100
y2 = df2['dxdt']

plt.plot(x, y, label="MuX=0.1")
plt.plot(x2, y2,label="MuX=0.2")
plt.xlabel("Time")
plt.ylabel("Cells Produced")
plt.legend()
```
